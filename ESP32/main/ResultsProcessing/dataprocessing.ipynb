```python
import pandas as pd
import numpy as np
import json
import plotly.express as px


# load the json data of data.txt
with open('data.txt') as json_file:
    data = json.load(json_file)

# "labels" is the list of the labels of the data
labels = data['labels']

# make a list with each one of the datasets in the data
data_list = []
for i in data["datasets"][2:]:
    data_list.append(i["data"])

data_names = []
for i in data["datasets"][2:]:
    data_names.append(i["label"])



# create a dataframe with each one of the datasets
df = pd.DataFrame(data_list)
# transpose the dataframe
df = df.transpose()
# add the labels to the dataframe
df.index = labels
# add the names of the datasets to the dataframe
df.columns = data_names

# delete de initial 10 values
df = df.iloc[20:]
# divide each value of each column by its first value
df = df.div(df.iloc[0]*10)

# create a gaussian filter
df= df.rolling(window=10).sum()


# plot the data
fig = px.line(df, x=df.index, y=df.columns)
fig.show()
```
